### Cleaning
Some articles are duplicated in our assets. These articles appear in the YF search for the ticker, and they usually describe macroeconomic situations. I decide to keep them in the dataset.

```python
df = df.drop(['Unnamed: 0.1','Unnamed: 0'],axis=1)
df = df[df.date != 'date_error']
df['date'] =  pd.to_datetime(df.date)
df.insert(0,'date',df.pop('date'))
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('data/articles_clean.csv',index=False)
```
